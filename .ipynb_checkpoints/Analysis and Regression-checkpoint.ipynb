#  RQ 2.  Examining the correlation between code quality, signified by cyclo-matic complexity, and 10x engineers.

```python
def simple_linear_regression(X, y):
    XX = X.to_numpy().reshape(-1, 1)
    ols = sm.OLS(y, X).fit()
    return ols, ols.summary()
```

```python
# Linear regression on cyclomatic complexity

X = df["average dmm"]
y = df["10x label churn"]

ols, summary = simple_linear_regression(X, y)
print(summary)

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(summary), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
plt.axis('off')
plt.tight_layout()
plt.savefig('outputs/output1.png')
```

```python
# Linear regression on self descriptiveness

X = df["average comments"]
y = df["10x label churn"]
ols, summary = simple_linear_regression(X, y)
print(summary)

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(summary), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
plt.axis('off')
plt.tight_layout()
plt.savefig('outputs/output2.png')
```

# Multiple Linear Regression

```python
X = df[['num owned files', 'average dmm', 'average comments']]
y = df["10x label churn"]

# standardize coef
df_z = X.select_dtypes(include=[np.number]).dropna().apply(stats.zscore)
df_z = sm.add_constant(df_z)

ols = sm.OLS(y, df_z).fit()
print(ols.summary())

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(ols.summary()), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
plt.axis('off')
plt.tight_layout()
plt.savefig('outputs/output4.png')
```
